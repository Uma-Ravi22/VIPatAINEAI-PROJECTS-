### Game Rules 
- 2 Players are required to play this game. The game will be played like this. For example, If Player1 chooses "ROCK" and 
- Player2 chooses "Scissors", then, according to the game rule, "ROCK CRUSHES SCISSORS". Hence ROCK is the WINNER. 

<b>Similarly the rule for all the choices between the 2 players is listed below.<b>

 
- Rock vs Paper      - Paper COVERS Rock           - Paper Wins.
- Rock vs Scissors   - Rock CRUSHES Scissors       - Rock Wins.
- Rock vs Spock      - Spock VAPORIZES Rock        - Spock wins.
- Rock vs Lizard     - Rock CRUSHES Lizard         - Rock Wins.
- Paper vs Scissors  - Scissors CUTS Paper         - Scissors Wins.
- Paper vs Spock     - Paper DISPROVES Spock       - Paper Wins.
- Paper vs Lizard    - Lizard EATS Paper           - Lizard Wins.
- Scissors vs Spock  - Spock SMASHES Scissors      - Spock Wins.
- Scissors vs Lizard - Scissors DECAPITATES Lizard - Scissors Wins.
- Lizard vs Spock    - Lizard POISONS Spock        - Lizard Wins.
